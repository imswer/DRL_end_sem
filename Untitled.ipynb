```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

class CouplingLayer(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(CouplingLayer, self).__init__()

        self.split = input_size // 2
        self.scale_net = nn.Sequential(
            nn.Linear(self.split, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, self.split),
            nn.Tanh()
        )

        self.translation_net = nn.Sequential(
            nn.Linear(self.split, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, self.split)
        )

    def forward(self, x, reverse=False):
        x1, x2 = torch.chunk(x, 2, dim=1)

        if not reverse:
            scale = self.scale_net(x2)
            translation = self.translation_net(x2)
            y1 = x1 * torch.exp(scale) + translation
        else:
            scale = -self.scale_net(x2)
            translation = -self.translation_net(x2)
            y1 = (x1 - translation) * torch.exp(scale)

        return torch.cat([y1, x2], dim=1)

class InvertibleCNN(nn.Module):
    def __init__(self, num_coupling_layers, input_size, hidden_size):
        super(InvertibleCNN, self).__init__()

        self.layers = nn.ModuleList([
            CouplingLayer(input_size, hidden_size) for _ in range(num_coupling_layers)
        ])

    def forward(self, x, reverse=False):
        batch_size = x.size(0)
        if reverse:
            x1, x2 = torch.chunk(x, 2, dim=1)
            x = torch.cat([x2, x1], dim=1)

        for layer in self.layers:
            x = layer(x, reverse)

        if reverse:
            x1, x2 = torch.chunk(x, 2, dim=1)
            x = torch.cat([x2, x1], dim=1)

        return x

# Example usage with a 10x10 matrix
num_coupling_layers = 2
input_size = 100  # Assuming a 10x10 matrix
hidden_size = 60

inn = InvertibleCNN(num_coupling_layers, input_size, hidden_size)

# Random input matrix
x = torch.randn(1, input_size)

# Forward pass
outputs = inn(x)

# Reverse pass
reconstructed_inputs = inn(outputs, reverse=True)

# Convert tensors to numpy arrays
x_np = x.view(10, 10).detach().numpy()
reconstructed_inputs_np = reconstructed_inputs.view(10, 10).detach().numpy()
x1_np, x2_np = torch.chunk(x, 2, dim=1)
x1_np = x1_np.view(5, 10).detach().numpy()
x2_np = x2_np.view(5, 10).detach().numpy()

# Plot original matrix
plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.title("Original Matrix")
plt.imshow(x_np, cmap='viridis')
plt.colorbar()

# Plot reconstructed matrix
plt.subplot(2, 3, 2)
plt.title("Reconstructed Matrix")
plt.imshow(reconstructed_inputs_np, cmap='viridis')
plt.colorbar()

# Plot split matrix 1
plt.subplot(2, 3, 4)
plt.title("Split Matrix 1")
plt.imshow(x1_np, cmap='viridis')
plt.colorbar()

# Plot split matrix 2
plt.subplot(2, 3, 5)
plt.title("Split Matrix 2")
plt.imshow(x2_np, cmap='viridis')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

class CouplingLayer(nn.Module):
    def __init__(self, in_channels, hidden_channels):
        super(CouplingLayer, self).__init__()

        # Split the input into two parts
        self.split = in_channels // 2

        # Create sub-networks for each part
        self.scale = nn.Sequential(
            nn.Conv2d(self.split, hidden_channels, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_channels, self.split, kernel_size=3, padding=1),
            nn.Tanh()
        )

        self.translation = nn.Sequential(
            nn.Conv2d(self.split, hidden_channels, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_channels, self.split, kernel_size=3, padding=1)
        )

    def forward(self, x, reverse=False):
        x1, x2 = torch.chunk(x, 2, dim=1)

        if not reverse:
            y1 = x1 * torch.tanh(self.scale(x2)) + self.translation(x2)
        else:
            y1 = (x1 - self.translation(x2)) * torch.tanh(self.scale(x2))

        return torch.cat([y1, x2], dim=1)

class InvertibleCNN(nn.Module):
    def __init__(self, num_coupling_layers, in_channels, hidden_channels):
        super(InvertibleCNN, self).__init__()

        self.layers = nn.ModuleList([
            CouplingLayer(in_channels, hidden_channels) for _ in range(num_coupling_layers)
        ])

    def forward(self, x, reverse=False):
        for layer in self.layers:
            x = layer(x, reverse)
        return x

# Define edge detection transformation
def detect_edges(image):
    edge_detected = image.filter(ImageFilter.FIND_EDGES)
    return edge_detected

# Example usage with a single image
num_coupling_layers = 1
in_channels = 3  # Assuming RGB images
hidden_channels = 64

inn = InvertibleCNN(num_coupling_layers, in_channels, hidden_channels)

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),  # Example augmentation
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Example augmentation
    transforms.ToTensor()
])

# Load image data using PIL and apply transformations
image_path = 'mm.jpg'
image = Image.open(image_path).convert('RGB')
image = transform(image).unsqueeze(0)  # Add batch dimension

# Forward pass
outputs = inn(image)
print("Forward pass output shape:", outputs.shape)

# Reverse pass
reconstructed_inputs = inn(outputs, reverse=True)
print("Reconstructed input shape:", reconstructed_inputs.shape)

# Convert torch tensors to numpy arrays
original_image = image[0].permute(1, 2, 0).numpy()
reconstructed_image = reconstructed_inputs[0].permute(1, 2, 0).detach().numpy()

# Edge detection on reconstructed image
edge_detected = detect_edges(Image.fromarray((reconstructed_image * 255).astype('uint8')))

# Display the images
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Reconstructed Image")
plt.imshow(reconstructed_image)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Edge Detected")
plt.imshow(edge_detected)
plt.axis("off")

plt.show()
```
